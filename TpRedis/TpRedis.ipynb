```python
# 1. On crée la clé 'User' avec la valeur "TonNom"
!redis-cli SET User "Etudiant"

# 2. On la renomme en 'User:1'
!redis-cli RENAME User User:1

# 3. On vérifie la nouvelle clé
!redis-cli GET User:1

# Création des autres utilisateurs
!redis-cli SET User:2 "Thomas"
!redis-cli SET User:3 "Sarah"
!redis-cli SET User:4 "Julie"

# Afficher TOUTES les clés qui existent
print("--- Toutes les clés ---")
!redis-cli KEYS *

# Afficher UNIQUEMENT celles qui commencent par "User"
print("--- Clés filtrées ---")
!redis-cli KEYS "User:*"

!redis-cli MGET User:1 User:2 User:3

# On ajoute " - Admin" à la fin de la valeur de User:1
!redis-cli APPEND User:1 " - Admin"

# On vérifie le résultat
!redis-cli GET User:1

# On essaie de créer User:1 (qui existe déjà !) avec l'option NX
!redis-cli SET User:1 "Hacker" NX

# On vérifie si la valeur a changé
!redis-cli GET User:1

import time

# 1. Créer une clé qui meurt dans 5 secondes
!redis-cli SET User:Temp "Ephémère" EX 5

# 2. Vérifier le temps restant tout de suite
print("Temps restant (secondes) :")
!redis-cli TTL User:Temp

# 3. On attend 6 secondes avec Python...
print("On attend 6 secondes...")
time.sleep(6)

# 4. On vérifie à nouveau
print("La valeur est-elle encore là ?")
!redis-cli GET User:Temp

!redis-cli SET User:6 "Paul"
!redis-cli SET User:6:City "Paris"
!redis-cli SET User:6:Age "30"
# Celle-ci expirera (ex: statut de connexion)
!redis-cli SET User:6:Activity "En ligne" EX 60

!redis-cli KEYS "User:6*"

# User:1 existe déjà. Donc User:10 ne devrait PAS être créé.
!redis-cli MSETNX User:10 "Nouveau" User:1 "Ecrasé?"

# Vérification
print("Est-ce que User:10 existe ?")
!redis-cli GET User:10

# Création du profil en une ligne
!redis-cli HSET User:Hash:1 nom "John" age 25 city "Lyon"

# Lecture de tout l'objet
!redis-cli HGETALL User:Hash:1

# Lecture juste du nom
!redis-cli HGET User:Hash:1 nom

# On envoie tout le bloc transactionnel d'un coup
!echo -e "MULTI\nLPUSH MaListe Tache1\nLPUSH MaListe Tache2\nLRANGE MaListe 0 -1\nEXEC" | redis-cli
```
